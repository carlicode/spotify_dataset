### 4. ¿Qué artista creó la canción más popular titulada "Love"?

#### Elimina todos los valores faltantes para responder al siguiente conjunto de preguntas.

Si el dataframe contiene valores faltantes, puedes eliminarlos utilizando el siguiente método:

`df.dropna(inplace=True)`

Este método busca elementos faltantes en el dataframe y, si encuentra alguno, elimina toda la fila correspondiente a la celda faltante.

Puedes encontrar más información en la documentación

### 7. Analisis del tempo de las canciones

Utilizar el comando `print(f"...")` donde corresponda para responder a las siguientes preguntas

Para los artistas con una popularidad mayor a 90 y cuyas canciones han sido lanzadas despues del 2015, encontrar la siguiente informacion:

- ¿Cuál es la canción más rápida en términos de tempo y en qué año fue lanzada?
- ¿Cuál es la canción más lenta en términos de tempo y en qué año fue lanzada?
- ¿Graficar el numero de canciones que fueron lanzadas por año y identificar cuál fue año con más canciones?


```python
# tu código aquí
nuevo_df = df.loc[(df['popularity'] > 90) & (df['year'] > 2015)]
#print("Nuevo DataFrame:")
#print(nuevo_df)
indice_max_tempo = df['tempo'].idxmax()
cancion_mas_rapida = df['name'].iloc[indice_max_tempo]
ano_lanzamiento = df['year'].iloc[indice_max_tempo]
print("Canción con el mayor tempo:", cancion_mas_rapida)
print("Año de lanzamiento:", ano_lanzamiento)

indice_min_tempo = df['tempo'].idxmin()
cancion_mas_lenta = df['name'].iloc[indice_min_tempo]
ano_lanzamiento = df['year'].iloc[indice_min_tempo]
print("Canción con el menor tempo:", cancion_mas_lenta)
print("Año de lanzamiento:", ano_lanzamiento)

```

```python
import matplotlib.pyplot as plt
conteo_por_ano = df['year'].value_counts()
conteo_por_ano.plot(kind='bar')
plt.xlabel('Año')
plt.ylabel('Número de Canciones')
plt.title('Número de Canciones por Año')
plt.show()

ano_con_mas_canciones = conteo_por_ano.idxmax()

print("Año con más canciones:", ano_con_mas_canciones)
```

### 8. Analisis del tiempo de duracion de las canciones

### 9. ¿Cuál es la canción más enérgica de Michael Jackson?

### 10 ¿Qué artista ha lanzado la mayor cantidad de canciones desde 1960?


```python
# tu código aquí
canciones_desde_1960 = df[df['year'] >= 1960]
conteo_por_artista = canciones_desde_1960['artist'].value_counts()
artista_con_mas_canciones = conteo_por_artista.idxmax()

print("Artista con mayor número de canciones desde 1960:", artista_con_mas_canciones)
```

### 11 Para el artista de la pregunta anterior, ¿cuál es la canción más popular?

```python
# tu código aquí
artista_mas_canciones = 'Bob Dylan'
canciones_artista_seleccionado = df[(df['artist'] == artista_mas_canciones) & (df['year'] >= 1960)]
canciones_ordenadas_por_popularidad = canciones_artista_seleccionado.sort_values(by='popularity', ascending=False)
cancion_mas_popular = canciones_ordenadas_por_popularidad.iloc[0]
print("Canción más popular del artista:", cancion_mas_popular['name'])
```

### 15 ¿Cuál es la distribución de la danceability de las canciones en función del año de lanzamiento? 
Crea un gráfico de barras para mostrar esta distribución.

```python
# tu código aquí
promedio_danceability_por_año = df.groupby('year')['danceability'].mean()
plt.figure(figsize=(10, 6))
promedio_danceability_por_año.plot(kind='bar', color='blue')
plt.title('Distribución de Danceability por Año')
plt.xlabel('year')
plt.ylabel('Danceability Promedio')
plt.grid(True)
plt.show()
```

### 16 ¿Cómo varía el tempo de las canciones a lo largo de los años?
Representa esta variación utilizando un gráfico de línea.


```python
# tu código aquí
promedio_tempo_por_año = df.groupby('year')['tempo'].mean()
plt.figure(figsize=(10, 6))
promedio_tempo_por_año.plot(kind='line', marker='o', color='green')
plt.title('Distribución de Tempo por Año')
plt.xlabel('Año')
plt.ylabel('Tempo Promedio')
plt.grid(True)
plt.show()
```

### 19  ¿Cuál es la energía promedio de todas las canciones de 'Michael Jackson'?
Para obtener el promedio de los valores en una columna de tipo numerico, puede utilizar el metodo

`df["nombre de la columna"].mean()`

```python
# tu código aquí
canciones_michael_jackson = df[df['artist'] == 'Michael Jackson']
promedio_energia_michael_jackson = canciones_michael_jackson['energy'].mean()
print("Promedio de energía para las canciones de Michael Jackson:", promedio_energia_michael_jackson)
```
