- ¬øCu√°l es la canci√≥n m√°s lenta en t√©rminos de tempo y en qu√© a√±o fue lanzada?

```python
# tu c√≥digo aqu√≠
nuevo_df = df.loc[(df['popularity'] > 90) & (df['year'] > 2015)]
#print("Nuevo DataFrame:")
#print(nuevo_df)
indice_max_tempo = df['tempo'].idxmax()
cancion_mas_rapida = df['name'].iloc[indice_max_tempo]
ano_lanzamiento = df['year'].iloc[indice_max_tempo]
print("Canci√≥n con el mayor tempo:", cancion_mas_rapida)
print("A√±o de lanzamiento:", ano_lanzamiento)

indice_min_tempo = df['tempo'].idxmin()
cancion_mas_lenta = df['name'].iloc[indice_min_tempo]
ano_lanzamiento = df['year'].iloc[indice_min_tempo]
print("Canci√≥n con el menor tempo:", cancion_mas_lenta)
print("A√±o de lanzamiento:", ano_lanzamiento)

```

```python
import matplotlib.pyplot as plt
conteo_por_ano = df['year'].value_counts()
conteo_por_ano.plot(kind='bar')
plt.xlabel('A√±o')
plt.ylabel('N√∫mero de Canciones')
plt.title('N√∫mero de Canciones por A√±o')
plt.show()

ano_con_mas_canciones = conteo_por_ano.idxmax()

print("A√±o con m√°s canciones:", ano_con_mas_canciones)
```

### 10 ¬øQu√© artista ha lanzado la mayor cantidad de canciones desde 1960?

### 15 ¬øCu√°l es la distribuci√≥n de la danceability de las canciones en funci√≥n del a√±o de lanzamiento? 
Crea un gr√°fico de barras para mostrar esta distribuci√≥n.

```python
# tu c√≥digo aqu√≠
promedio_danceability_por_a√±o = df.groupby('year')['danceability'].mean()
plt.figure(figsize=(10, 6))
promedio_danceability_por_a√±o.plot(kind='bar', color='blue')
plt.title('Distribuci√≥n de Danceability por A√±o')
plt.xlabel('year')
plt.ylabel('Danceability Promedio')
plt.grid(True)
plt.show()
```

### 16 ¬øC√≥mo var√≠a el tempo de las canciones a lo largo de los a√±os?
Representa esta variaci√≥n utilizando un gr√°fico de l√≠nea.


```python
# tu c√≥digo aqu√≠
promedio_tempo_por_a√±o = df.groupby('year')['tempo'].mean()
plt.figure(figsize=(10, 6))
promedio_tempo_por_a√±o.plot(kind='line', marker='o', color='green')
plt.title('Distribuci√≥n de Tempo por A√±o')
plt.xlabel('A√±o')
plt.ylabel('Tempo Promedio')
plt.grid(True)
plt.show()
```

### 19  ¬øCu√°l es la energ√≠a promedio de todas las canciones de 'Michael Jackson'?
Para obtener el promedio de los valores en una columna de tipo numerico, puede utilizar el metodo

`df["nombre de la columna"].mean()`

```python
# tu c√≥digo aqu√≠
canciones_michael_jackson = df[df['artist'] == 'Michael Jackson']
promedio_energia_michael_jackson = canciones_michael_jackson['energy'].mean()
print("Promedio de energ√≠a para las canciones de Michael Jackson:", promedio_energia_michael_jackson)
```
